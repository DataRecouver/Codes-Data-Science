# 1- Utilizando Regressão Linear:

```python
df_newdata.fillna(method ='ffill', inplace = True)
```

```python

#features = df_newdata[['age', 'occupation_code']]
#print(features)
```

```python
##TESTE
# Separando os dados em features e alvo (variável a ser prevista)
#X = features
#y = df_newdata['weeks worked in year']

# Criando o modelo
#modelo = LinearRegression()

# Treinando o modelo
#modelo.fit(X, y)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
X = np.array(df_newdata['age']).reshape(-1, 1)

y = np.array(df_newdata['occupation_code']).reshape(-1, 1)

# # Separando os dados em variáveis ​​independentes e dependentes

# # Convertendo cada dataframe em um array numpy 

# Já que cada dataframe contém apenas uma coluna

df_newdata.dropna(inplace = True)

# # Descartando qualquer linha com valores Nan

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5)

# # Dividindo os dados em dados de treinamento e teste

regr = LinearRegression()

regr.fit(X_train, y_train)

print(regr.score(X_test, y_test))
```

Pode ser visto que o modelo não apresenta alta precisão, sendo necessário mais testes.

```python
y_pred = regr.predict(X_test)

plt.scatter(X_test, y_test, color ='b')

plt.plot(X_test, y_pred, color ='k')

plt.show()
# Dispersão de dados de valores previstos
```

## 2-Utilizando Random Forrest

```python
from sklearn.ensemble import RandomForestClassifier
random= RandomForestClassifier(max_depth=2, random_state=0)
random.fit(X_train, y_train)
y_pred = random.predict(X_test)
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, random_state=1, shuffle = True)
```

```python
#Teste de Acuracia não confiavel niveis muito baixos de precisão do modelo
from sklearn.metrics import classification_report, confusion_matrix  ,accuracy_score
Random_Forest_Accuracy=random.score(X_test, y_test)*100
print(f'The Accuracy of the Model is {Random_Forest_Accuracy}')
```

##3- Árvore de decisao

## 4- Utilizando Rede Neural

```python
from sklearn.neural_network import MLPClassifier
neural = MLPClassifier(random_state=1, max_iter=1000).fit(X_train, y_train)
y_pred = neural.predict(X_test)
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, random_state=1, shuffle =True)
```

### 10. Exiba o total de vendas de cada produto ao longo dos meses em um gráfico de área. Defina um grau de transparência ao gráfico para deixar a visualização mais agradável.

O gráfico deve ser semelhante a esse:

image.png

```
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots()

for column in ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]:
    ax.fill_between(sales_df.index+1, sales_df[column], label = column,alpha=0.5)
ax.set_xlabel("Mês")
ax.set_ylabel("Unidades Vendidas")
ax.legend(loc="upper left");
plt.xticks([i for i in range(1,13)])
```

### 11. Compare o volume de vendas de hidratante e creme facial com um gráfico de barras.

O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico. 

Neste mesmo gráfico, apresente o total de produtos vendidos em cada mês. Como a escala de quantidade de produtos é diferente da escala de lucro, crie um eixo gêmeo (pesquise por twin axes).




O gráico deve ser semelhante a esse:

image.png

```
fig, ax = plt.subplots()
x = 0
width = 0.2
for column in ["facecream","moisturizer"]:
    ax.bar(sales_df.index+1+x, sales_df[column], label = column,width = width)
    x+=width

ax2 = ax.twinx()
ax2.plot(sales_df.index+1+width/2, sales_df["total_profit"],linestyle='-',color = "green", linewidth=3, marker='o',markerfacecolor='white',
         markeredgecolor='red',markeredgewidth=1)
ax.set_xlabel("Mês")
ax.set_ylabel("Unidades Vendidas")
ax.legend(loc="upper left");
plt.xticks([i for i in range(1,13)])
```

### 12. Crie uma grade de subplots e, para cada painel, exiba um gráfico de dispersão relacionando os dados de vendas de cada produto e o total de lucro obtido no mês. Na última linha da grade deve ser apresentado um único painel com um gráfico de caixas exibindo a distribuição de vendas de cada produto. 


```
columns = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
fig = plt.figure(figsize=(18, 6))
grid = plt.GridSpec(2, 6, hspace=0.2, wspace=0.2)
fig.add_gridspec(3,3, hspace=10)
main_ax = fig.add_subplot(grid[1,:])
main_ax.boxplot(sales_df[columns].T,labels=columns);
for counter,column in enumerate(columns):
    if(counter==0):
        ax = fig.add_subplot(grid[0,counter])
        ax.plot( sales_df[column],sales_df['total_profit'],'ok', markersize=3)
        base_ax = ax
    else:
        ax = fig.add_subplot(grid[0,counter], yticklabels=[])
        ax.plot( sales_df[column],sales_df['total_profit'],'ok', markersize=3)
```
